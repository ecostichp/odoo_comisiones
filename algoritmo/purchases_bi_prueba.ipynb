```python
from wep import complete_df
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.float_format = '{:,.2f}'.format
```

```python
from wep import descrip_sales_users_df, desc_product_df
```

```python
ventas = complete_df[['salesman_id', 'product_id','quantity', 'price_subtotal']].merge(descrip_sales_users_df, how='left', left_on='salesman_id', right_on='id')
compras = ventas.merge(desc_product_df, how='left', left_on='product_id', right_on='id_producto')
compras['id_producto'] = compras['id_producto'].astype('Int64')
```
