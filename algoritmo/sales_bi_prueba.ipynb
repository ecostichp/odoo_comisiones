```python
from wep import complete_df
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.float_format = '{:,.2f}'.format
```

```python
from wep import descrip_sales_users_df
```

```python
mes = complete_df[['salesman_id', 'quantity', 'price_subtotal']].groupby('salesman_id', dropna=False).sum()

ventas_mes = mes.reset_index(
).merge(
    descrip_sales_users_df, how='left', left_on='salesman_id', right_on='id'
).set_index(
    'salesman_id'
).drop(
    columns=['id', 'sale_team_id', 'state']
)

ventas_mes
```
