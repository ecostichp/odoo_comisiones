```python
cobranza_SAE_año_w_nc= pd.concat([cobranza_sae_CCC_año, cobranza_sae_CCV_año], ignore_index=True)

cobranza_SAE_año_merge = cobranza_SAE_año_w_nc[cobranza_SAE_año_w_nc['Descripción'] != 'Nota de crédito'].merge(
                                descrip_sales_users_df, how='left', left_on='VendOdoo', right_on='salesperson_id').merge(
                                descrip_partner_df, how='left', left_on='Id Odoo', right_on='partner_id')

cobranza_SAE_año_merge['pay_journal'] = 'SAE'

cobranza_SAE_año_sorted = cobranza_SAE_año_merge.sort_values(['Fecha de aplicación', 'Referencia', 'business_model'])


cobranza_SAE_año_descript = cobranza_SAE_año_sorted[['Referencia', 'partner_id', 'partner_name', 'Fecha Fact',
                                                    'pay_journal', 'Importe', 'Fecha de aplicación',
                                                    'salesperson_id', 'salesperson_name', 'sale_team_id',
                                                    'sale_team_description', 'business_model', 'warehouse', 'Dias']]

cobranza_SAE_año_descript.columns = ['fact_doc_name', 'partner_id', 'partner_name', 'fact_doc_date',
                                    'pay_journal', 'pay_amount', 'pay_date',
                                    'salesperson_id', 'salesperson_name', 'sale_team_id',
                                    'sale_team_description', 'business_model', 'warehouse', 'dias_pago']
```

```python
cobranza_año_descript_before_observaciones = pd.concat([cobranza_odoo_año_descript, cobranza_SAE_año_descript], ignore_index=True)

cobranza_año_descript_before_observaciones['fact_doc_id'] = cobranza_año_descript_before_observaciones['fact_doc_id'].astype('Int64')
```

```python
d1, d2, d3, d4, d5, d6 = 15, 30, 45, 60, 75, 90

cobranza_año_descript_before_observaciones['pago_0_15'] = cobranza_año_descript_before_observaciones['dias_pago'] <= d1
cobranza_año_descript_before_observaciones['pago_16_30'] = (cobranza_año_descript_before_observaciones['dias_pago'] > d1) & (cobranza_año_descript_before_observaciones['dias_pago'] <= d2)
cobranza_año_descript_before_observaciones['pago_31_45'] = (cobranza_año_descript_before_observaciones['dias_pago'] > d2) & (cobranza_año_descript_before_observaciones['dias_pago'] <= d3)
cobranza_año_descript_before_observaciones['pago_46_60'] = (cobranza_año_descript_before_observaciones['dias_pago'] > d3) & (cobranza_año_descript_before_observaciones['dias_pago'] <= d4)
cobranza_año_descript_before_observaciones['pago_61_75'] = (cobranza_año_descript_before_observaciones['dias_pago'] > d4) & (cobranza_año_descript_before_observaciones['dias_pago'] <= d5)
cobranza_año_descript_before_observaciones['pago_76_90'] = (cobranza_año_descript_before_observaciones['dias_pago'] > d5) & (cobranza_año_descript_before_observaciones['dias_pago'] <= d6)
cobranza_año_descript_before_observaciones['pago_91_mas'] = cobranza_año_descript_before_observaciones['dias_pago'] > d6
```

```python
file03 = 'observaciones_cobranza'
file03_path_str = str(data_projetc_path.joinpath(f'cobranza/{file03}.xlsx'))

observaciones_cobranza = pd.read_excel(file03_path_str, dtype_backend='numpy_nullable')

observaciones_cobranza['cobranza_autorizada'] = True
```

```python
check1 = cobranza_año_descript_before_observaciones.merge(
                            observaciones_cobranza,
                            how='right',
                            on='fact_doc_name'
                        )

if not check1[check1['partner_id'].isna()].empty:
    print('Error, algunas observaciones no encontraron un nombre de factura para empatarse. Corrígelas!')
    display(check1.loc[check1['partner_id'].isna(), 'fact_doc_name'])

else:
    cobranza_año_descript = cobranza_año_descript_before_observaciones.merge(
                        observaciones_cobranza,
                        how='left',
                        on='fact_doc_name'
                    )
    
    cobranza_año_descript['paga_comision'] = cobranza_año_descript['pago_91_mas'] == False 
    cobranza_año_descript.loc[cobranza_año_descript['cobranza_autorizada'] == True, 'paga_comision'] = True
```
