```python
from sys import path
from pathlib import Path
from sqlalchemy import create_engine

import pandas as pd
import numpy as np
pd.options.display.float_format = '{:,.2f}'.format

# Se genera la ruta absoluta al directorio "Comisiones"
path_package = str(Path.cwd().parent)

# Se agrega la ruta al directorio sys.path
path.append(path_package)

# Se define la ruta para la base de datos local SQLite
db_file = 'comisiones.db'
db_file_path_str = str(Path().cwd().parent.parent.joinpath(f'data/{db_file}'))

# Se crea el objeto engine para trabajarlo con los metodos de pandas
engine = create_engine(f'sqlite:///{db_file_path_str}')


from algoritmos.algoritmo_cobranza_mes import cobranza_mes_func
from descriptions.descriptions import descrip_sales_users_df, descrip_partner_df
```

```python
with engine.connect() as conn, conn.begin():  
    ventas_año = pd.read_sql_table('ventas_año', conn, dtype_backend='numpy_nullable')

engine.dispose()
```

```python
cobranza_odoo_enero = cobranza_mes_func(1)
cobranza_odoo_febrero = cobranza_mes_func(2)
cobranza_odoo_marzo = cobranza_mes_func(3)
cobranza_odoo_abril = cobranza_mes_func(4)
cobranza_odoo_mayo = cobranza_mes_func(5)
cobranza_odoo_junio = cobranza_mes_func(6)
# cobranza_odoo_julio = cobranza_mes_func(7)

cobranza_odoo_año = pd.concat([
                        cobranza_odoo_enero, 
                        cobranza_odoo_febrero, 
                        cobranza_odoo_marzo, 
                        cobranza_odoo_abril, 
                        cobranza_odoo_mayo,
                        cobranza_odoo_junio,
                        # cobranza_odoo_julio,
                        ], 
                        ignore_index=True)

cobranza_odoo_año['dias_pago'] = (cobranza_odoo_año['pay_date'] - cobranza_odoo_año['fact_doc_date']).dt.days
```

```python
cobranza_odoo_año_descript = (cobranza_odoo_año.merge(
              ventas_año[['fact_doc_id', 'salesperson_id', 'salesperson_name', 'sale_team_id', 'sale_team_description', 
                    'business_model', 'warehouse']].drop_duplicates('fact_doc_id'), how='left', on='fact_doc_id')).merge(
                     descrip_partner_df, how='left', left_on= 'fact_doc_cliente', right_on= 'partner_id'
                     )[[
                     'fact_doc_id', 'fact_doc_name', 'fact_doc_cliente', 'partner_name', 'fact_doc_date', 'fact_doc_total', 
                     'fact_doc_deuda', 'pay_journal', 'ref', 'pay_amount', 'pay_date', 'dias_pago', 'pay_pos', 'salesperson_id', 'salesperson_name', 
                     'sale_team_id', 'sale_team_description', 'business_model', 'warehouse']]


cobranza_odoo_año_descript.columns = ['fact_doc_id', 'fact_doc_name', 'partner_id', 'partner_name',
       'fact_doc_date', 'fact_doc_total', 'fact_doc_deuda', 'pay_journal',
       'ref', 'pay_amount', 'pay_date', 'dias_pago', 'pay_pos', 'salesperson_id',
       'salesperson_name', 'sale_team_id', 'sale_team_description',
       'business_model', 'warehouse']
```

```python
cobranza_SAE_año_w_nc= pd.concat([cobranza_sae_CCC_año, cobranza_sae_CCV_año], ignore_index=True)

cobranza_SAE_año_merge = cobranza_SAE_año_w_nc[cobranza_SAE_año_w_nc['Descripción'] != 'Nota de crédito'].merge(
                                descrip_sales_users_df, how='left', left_on='VendOdoo', right_on='salesperson_id').merge(
                                descrip_partner_df, how='left', left_on='Id Odoo', right_on='partner_id')

cobranza_SAE_año_merge['pay_journal'] = 'SAE'

cobranza_SAE_año_sorted = cobranza_SAE_año_merge.sort_values(['Fecha de aplicación', 'Referencia', 'business_model'])

cobranza_SAE_año_sorted['id'] = [-i -1 for i in range(len(cobranza_SAE_año_sorted))]


cobranza_SAE_año_descript = cobranza_SAE_año_sorted[['id', 'Referencia', 'partner_id', 'partner_name', 'Fecha Fact',
                                                    'pay_journal', 'Importe', 'Fecha de aplicación', 'Dias',
                                                    'salesperson_id', 'salesperson_name', 'sale_team_id',
                                                    'sale_team_description', 'business_model', 'warehouse']]

cobranza_SAE_año_descript.columns = ['fact_doc_id', 'fact_doc_name', 'partner_id', 'partner_name', 'fact_doc_date',
                                    'pay_journal', 'pay_amount', 'pay_date', 'dias_pago',
                                    'salesperson_id', 'salesperson_name', 'sale_team_id',
                                    'sale_team_description', 'business_model', 'warehouse']
```

```python
cobranza_año_descript = pd.concat([cobranza_odoo_año_descript, cobranza_SAE_año_descript], ignore_index=True)

cobranza_enero_descript = cobranza_año_descript.loc[cobranza_año_descript['pay_date'].dt.month == 1]
cobranza_febrero_descript = cobranza_año_descript.loc[cobranza_año_descript['pay_date'].dt.month == 2]
cobranza_marzo_descript = cobranza_año_descript.loc[cobranza_año_descript['pay_date'].dt.month == 3]
cobranza_abril_descript = cobranza_año_descript.loc[cobranza_año_descript['pay_date'].dt.month == 4]
cobranza_mayo_descript = cobranza_año_descript.loc[cobranza_año_descript['pay_date'].dt.month == 5]
cobranza_junio_descript = cobranza_año_descript.loc[cobranza_año_descript['pay_date'].dt.month == 6]
# cobranza_julio_descript = cobranza_año_descript.loc[cobranza_año_descript['pay_date'].dt.month == 7]
```

```python
with engine.connect() as conn, conn.begin():  
    cobranza_enero_descript.to_sql(name = 'cobranza_enero', index= False, con = conn, if_exists = 'replace')
    cobranza_febrero_descript.to_sql(name = 'cobranza_febrero', index= False, con = conn, if_exists = 'replace')
    cobranza_marzo_descript.to_sql(name = 'cobranza_marzo', index= False, con = conn, if_exists = 'replace')
    cobranza_abril_descript.to_sql(name = 'cobranza_abril', index= False, con = conn, if_exists = 'replace')
    cobranza_mayo_descript.to_sql(name = 'cobranza_mayo', index= False, con = conn, if_exists = 'replace')
    cobranza_junio_descript.to_sql(name = 'cobranza_junio', index= False, con = conn, if_exists = 'replace')
    # cobranza_julio_descript.to_sql(name = 'cobranza_julio', index= False, con = conn, if_exists = 'replace')
    cobranza_año_descript.to_sql(name = 'cobranza_año', index= False, con = conn, if_exists = 'replace')

engine.dispose()
```
