```python
import xmlrpc.client
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

api_url = os.environ.get('ODOO_URL_API')
api_db = os.environ.get('ODOO_DB_API')
api_username = os.environ.get('ODOO_USERNAME_API')
api_clave = os.environ.get('ODOO_CLAVE_API')

common = xmlrpc.client.ServerProxy(f'{api_url}/xmlrpc/2/common')
uid = common.authenticate(api_db, api_username, api_clave, {})
models = xmlrpc.client.ServerProxy(f'{api_url}/xmlrpc/2/object')
```

```python
account_move_line_fields = models.execute_kw(api_db, uid, api_clave, 'account.move.line', 'fields_get', [])
for i in account_move_line_fields.keys():
    try:
        models.execute_kw(api_db, uid, api_clave, 'account.move.line', 'read', [126870], {'fields': [i]})
    except:
        print(i)
```

```python
list_fields = list(account_move_line_fields.keys())
list_fields.remove('compute_all_tax')
```

```python
ids_fac_doc = models.execute_kw(api_db, uid, api_clave, 'account.move', 'search', [[("name", "=", "F1-CC/2024/03156")]])
fac_doc = models.execute_kw(api_db, uid, api_clave, 'account.move', 'read', [ids_fac_doc])
fac_doc
```

```python
ids_fac_line = fac_doc[0]['line_ids']
fac_line = models.execute_kw(api_db, uid, api_clave, 'account.move.line', 'read', [ids_fac_line], {'fields': list_fields})
fac_line
```

```python
ids_pos_doc = fac_doc[0]['pos_order_ids']
ids_pos_line = models.execute_kw(api_db, uid, api_clave, 'pos.order.line', 'search', [[("order_id.id", "=", ids_pos_doc)]])
pos_line = models.execute_kw(api_db, uid, api_clave, 'pos.order.line', 'read', [ids_pos_line])
pos_line
```

```python
sale_order_line_id = pos_line[0]['sale_order_line_id'][0]
sale_line = models.execute_kw(api_db, uid, api_clave, 'sale.order.line', 'read', [sale_order_line_id])
sale_line
```
