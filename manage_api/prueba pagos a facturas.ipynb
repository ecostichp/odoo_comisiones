```python
import xmlrpc.client
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

api_url = os.environ.get('ODOO_URL_API')
api_db = os.environ.get('ODOO_DB_API')
api_username = os.environ.get('ODOO_USERNAME_API')
api_clave = os.environ.get('ODOO_CLAVE_API')

common = xmlrpc.client.ServerProxy(f'{api_url}/xmlrpc/2/common')
uid = common.authenticate(api_db, api_username, api_clave, {})
models = xmlrpc.client.ServerProxy(f'{api_url}/xmlrpc/2/object')
```

```python
[id1] = models.execute_kw(api_db, uid, api_clave, 'account.move', 'search', [[("name", "=", "F1-CC/2024/02702")]])
[id2] = models.execute_kw(api_db, uid, api_clave, 'account.move', 'search', [[("name", "=", "F1-VS/2024/00301")]])
[id3] = models.execute_kw(api_db, uid, api_clave, 'account.move', 'search', [[("name", "=", "F1-VS/2024/00713")]])
[id4] = models.execute_kw(api_db, uid, api_clave, 'account.move', 'search', [[("name", "=", "F1-CC/2024/02702")]])
[id5] = models.execute_kw(api_db, uid, api_clave, 'account.move', 'search', [[("name", "=", "F1-CC/2024/01961")]])

ids = [id1, id2, id3, id5]
```

```python
fields = [
    'name',
    'partner_id',
    'date',
    'amount_total',
    'amount_residual',
    'invoice_has_outstanding',
    'invoice_payments_widget',
]

wep = models.execute_kw(api_db, uid, api_clave, 'account.move', 'read', [ids], {'fields':fields})
wep
```
